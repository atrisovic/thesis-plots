# LHCb Monte Carlo and raw experimental data size ratios by the end of 2016




Source: https://lhcb-portal-dirac.cern.ch

```python
%run header
%matplotlib inline
wedgeprops = {'linewidth' : .5 , 'edgecolor' : 'grey'}
fontsize = 12
bar_width = 0.5
opacity = 0.6
```

# Total data accumulated by December 2016

```python
n_groups = 2
values = (22.4, 10.2)
names = ['Raw data', 'Monte Carlo']
y_pos = np.arange(len(values))
col = plt.cm.Set1(np.linspace(0,1,10))

bar_width = 0.35
opacity = 0.4
plt.bar(y_pos, values, bar_width, alpha=1, align='center', color=col)
plt.xticks(y_pos, names)
plt.ylabel('Size in petabytes (PB)')
plt.xlabel('Data type')
plt.title('Total LHCb data and Monte Carlo sample volume accumulated by December 2016', fontsize=14)
plt.show()
```

## Total raw data collected in Run 1 and in Run 2

```python
labels = 'Run 1', 'Run 2'
sizes = [6.5, 23.0]
explode = (0, 0.1)
colors = plt.cm.Set3(np.linspace(0,1,10))

fig = plt.figure()
fig.suptitle('LHCb Run 1 vs Run 2 data size', fontsize=14)
ax = fig.gca()

ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90, wedgeprops=wedgeprops, colors=colors)
ax.axis('equal')  

fig.tight_layout()
plt.show()

fig.savefig("comparison.pdf")
```

## Total amount of data for physics analyses (legacy stripping) in Run 1

Total amount of the dataset is 1 Petabyte. 50% of it is to be released as open data in 2018

```python
labels = '2012 Stripping', '2012 Incremental Stripping', '2011 Stripping', '2011 Incremental Stripping'
sizes = [49.1, 27.9, 15.3, 7.6]
explode = (0, 0, 0.1, 0.15)
colors = plt.cm.Set3(np.linspace(0,1,10))

fig = plt.figure()
fig.suptitle('LHCb Run 1 data ratio', fontsize=14)
ax = fig.gca()

ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=50, wedgeprops=wedgeprops, colors=colors)
ax.axis('equal')  

fig.tight_layout()
plt.show()
```

## Raw experimental data production

```python
labels1 = 'Collision 16', 'Collision 12', 'Collision 15', 'Ion Proton 16',\
'Collision 11', 'Proton Helium 16', 'Lead 15', 'Proton Ion 16', 'Collision 10', 'Other'
        
sizes = [43, 17, 13, 7, 6, 3, 3, 2, 2, 4]
explode = (0.1, 0, 0, 0, 0, 0.1, 0.2, 0.3, 0.4, 0.45)
colors = plt.cm.Set3(np.linspace(0,1,10))
y_pos = np.arange(len(sizes))

#plt.rcdefaults()
#fig, ax = plt.subplots()
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
fig = plt.figure()
#
#fig.title('LHCb Run 1 data ratio', fontsize=14)
ax = fig.gca()
ax.set_title('LHCb Raw data size ratios by the end of 2016', fontsize=14)

#ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
#        shadow=False, startangle=50, wedgeprops=wedgeprops, colors=colors)

ax.barh(y_pos, sizes, color=colors, edgecolor='k', linewidth=0.5)

ax.set_yticks(y_pos)
ax.set_yticklabels(labels1, fontsize=12)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Percentage (%) of the total raw data size', fontsize=12)

fig.tight_layout()
plt.show()

fig.savefig("raw_data.pdf")
```
