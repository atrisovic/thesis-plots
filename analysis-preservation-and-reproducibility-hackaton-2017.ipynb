# LHCb Survey Results

```python
# %load header
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
```

## Question: How often do you reproduce final datasets in your analysis?

```python
n_groups = 4

results = (43, 18, 31, 8)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 1

result = ax.bar(index + bar_width/2, results, bar_width,
                alpha=opacity, color='b')

ax.set_xlabel('Answers')
ax.set_ylabel('Scores (in %)')
ax.set_title('Scores by answer')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('Every few months', 'Every few weeks', 'Once at most', 'Other'))

fig.tight_layout()
plt.show()
```

## Do you have access to your colleagues' code?

```python
labels = 'Yes', 'Probably yes', 'No'

sizes = [46, 34, 20]
explode = (0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  

plt.show()
```
