# LHCb Survey Results

```python
# %load header
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
```

## Survey question: How often do you reproduce final datasets in your analysis?

```python
labels = 'Every few months', 'Every few weeks', 'Once at most', 'Other'

results = (43, 18, 31, 8)
explode = (0.1, 0.1, 0, 0)
colors = plt.cm.Set3(np.linspace(0,1,4))

fig1, ax1 = plt.subplots()
ax1.pie(results, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90, colors=colors)
ax1.axis('equal')  

plt.show()
```

## Survey question: Do you have access to your colleagues' code?

```python
labels = 'Yes', 'Probably yes', 'No'

sizes = [46, 34, 20]
explode = (0.1, 0, 0)
colors = plt.cm.Set3(np.linspace(0,1,3))

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90, colors=colors)
ax1.axis('equal')  

plt.show()
```

## Survey question: Do you have a code repository for your analysis?

```python
labels2 = 'Yes, on github', 'Yes, on SVN', 'No'

sizes2 = [80, 6, 14]
explode2 = (0.1, 0, 0)

fig2, ax2 = plt.subplots()
ax2.pie(sizes2, explode=explode, labels=labels2, autopct='%1.1f%%',
        shadow=False, startangle=90, colors=colors)
ax2.axis('equal')  

plt.show()
```

# Survey question: How familiar are you with ___ ?

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
fig.suptitle("How familiar are you with ...", fontsize=16)
n_groups = 4
options = ['Using it regularly', 'Customising it\nfor my purpose', 'Used it once or\ntwice', 'Never used']
y_pos = [1,2,3,4]
bar_width = 0.5
opacity = 0.6
fontsize = 12
col = plt.cm.PiYG(np.linspace(0,1,4))

# first box
ax1.bar(y_pos, (9, 4, 2, 1), bar_width, alpha=opacity, align='center', color=col)
ax1.set_ylabel('Frequency', fontsize=fontsize)
ax1.set_title('CERN Open stack', fontsize=fontsize)
ax1.set_xticks(y_pos)
ax1.set_xticklabels(options, rotation=0)

# second
ax2.set_title("Docker")
ax2.bar(y_pos, (11, 3, 2, 0), bar_width, alpha=opacity, align='center', color=col)
ax2.set_xticks(y_pos)
ax2.set_xticklabels(options, rotation=0)

# third
ax3.set_ylabel('Frequency', fontsize=fontsize)
ax3.set_title("make-like pipeline systems")
ax3.bar(y_pos, (7, 3, 2, 4), bar_width, alpha=opacity, align='center', color=col)
ax3.set_xticks(y_pos)
ax3.set_xticklabels(options, rotation=0)

# forth
ax4.set_title("continuous integration")
ax4.bar(y_pos, (10,0,3,1), bar_width, alpha=opacity, align='center', color=col)
ax4.set_xticks(y_pos)
ax4.set_xticklabels(options, rotation=0)

fig.tight_layout(pad=0.4, w_pad=0.5, h_pad=2.5)
fig.subplots_adjust(top=0.9)
plt.show()
```
